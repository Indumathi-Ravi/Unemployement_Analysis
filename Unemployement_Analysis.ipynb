```python
# Average unemployment rate by year
avg_unemployment_year = data[['2014', '2015', '2016', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']].mean()
plt.figure(figsize=(9,4))
avg_unemployment_year.plot(kind='line', marker='o')
plt.title('Average Unemployment Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Unemployment Rate %')
plt.show()
```

```python
# Boxplot of unemployment rate by gender
sns.boxplot(x='sex', y='2023', data=data)
plt.title('Unemployment Rate by Gender in 2023')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate %')
plt.show()
```

```python
# Compare unemployment rates across gender and age groups
plt.figure(figsize=(10,6))
sns.barplot(x='sex', y='2023', hue='age_group', data=data)
plt.title('Unemployment Rate by Gender and Age Group in 2023')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate %')
plt.show()
```

```python
# Compare unemployment rates between countries in a specific year
plt.figure(figsize=(25, 6))
sns.barplot(x='country_name', y='2023', data=data)
plt.title('Unemployment Rate Comparison Across Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Unemployment Rate %')
plt.xticks(rotation=90)
plt.show()
```

```python
# Compare unemployment rates between countries in a specific year
plt.figure(figsize=(10, 30))
sns.barplot(x='2023', y='country_name', data=data)
plt.title('Unemployment Rate Comparison Across Countries in 2023')
plt.xlabel('Unemployment Rate %')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Exclude non-numeric columns for correlation analysis
numeric_columns = data.select_dtypes(include='number').columns
correlation_matrix = data[numeric_columns].corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
